```python
#comparing the files of data.docs and New.docs
def compare_files():
    data_docs = 0
    dir = "C:/Users/Administrator/Desktop/Data/docs"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            data_docs += 1
    print("files in Data->docs",data_docs)

    new_docs = 0
    dir = "C:/Users/Administrator/Desktop/New/docs"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            new_docs += 1
    print("files in New->docs",new_docs)

    if data_docs==new_docs:
        print("Number of files are equal")
    else:
        print("Number of files are not equal")


    #comparing the files of data.pdf and New.pdf
    data_pdf = 0
    dir = "C:/Users/Administrator/Desktop/Data/pdf"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            data_pdf += 1
    print("files in Data->pdf",data_pdf)

    new_pdf = 0
    dir = "C:/Users/Administrator/Desktop/New/pdf"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            new_pdf += 1
    print("files in New->pdf",new_pdf)

    if data_pdf==new_pdf:
        print("Number of files are equal")
    else:
        print("Number of files are not equal")


    #comparing the files of data.jpg and New.jpg
    data_jpg = 0
    dir = "C:/Users/Administrator/Desktop/Data/jpg"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            data_jpg += 1
    print("files in Data->jpg",data_jpg)

    new_jpg = 0
    dir = "C:/Users/Administrator/Desktop/New/jpg"
    for path in os.listdir(dir):
        if os.path.isfile(os.path.join(dir, path)):
            new_jpg += 1
    print("files in New->jpg",new_jpg)

    if data_jpg==new_jpg:
        print("Number of files are equal")
    else:
        print("Number of files are not equal")
```

```python
#search a file
def search_files():
    flag=0
    path="C:/Users/Administrator/Desktop"
    file_name=input("Enter the name of the file you want to search : ")
    for root, dirs, files in os.walk(path):
        if file_name in files:
            print(os.path.join(root,file_name))
            print(file_name,"found\n")
            flag=1
    if flag==0:
        print(file_name,"not found")
```

```python
#file operations

def delete_files():
    file_name=input("Enter the name of the file you want to delete: ")
    path="C:/Users/Administrator/Desktop"
    flag=0
    for root, dirs, files in os.walk(path):
        try:
            if file_name in files:
                os.remove(os.path.join(root,file_name))
                print("file deleted\n")
                flag=1
            

        except FileNotFoundError:
            print("File not found\n")
    if flag==0:
                print("File not found\n")
```

```python
#coping files
import shutil
def copy_files():
   
    src_file=input("Enter file you want to copy")
    dest_dir=input("Enter folder name Data or New")
    dest_folder=input("Enter folder name docs/ pdf/jpg")
    path="C:/Users/Administrator/Desktop"
    destination=os.path.join(path,dest_dir)
    destination1=os.path.join(destination,dest_folder)
    for root, dirs, files in os.walk(path):
        if src_file in files:
            src=os.path.join(root,src_file)
            shutil.copyfile(src,destination1)
            print("File successfully copied\n")
```

```python
#move files
import os
import shutil
def move_files():
    try:
        src_file=input("Enter file you want to move")
        dest_dir=input("Enter folder name Data or New")
        dest_folder=input("Enter folder name docs or pdf or jpg")
        path="C:/Users/Administrator/Desktop"
        destination=os.path.join(path,dest_dir)
        destination1=os.path.join(destination,dest_folder)
        for root, dirs, files in os.walk(path):
            if src_file in files:
                src=os.path.join(root,src_file)
                shutil.move(src,destination1),
                print("file successfully moved\n")
    except:
        print("File already exists\n")
 
```

```python
#finding duplicates
def duplicate_files():
    flag=1
    data_files=[]
    new_files=[]
    path_data="C:/Users/Administrator/Desktop/Data"
    path_new="C:/Users/Administrator/Desktop/New"
    for root, dirs, files in os.walk(path_data):
        for file in files:
            data_files.append(file)
    for root, dirs, files in os.walk(path_new):
        for file in files:
            new_files.append(file)
    print(data_files)
    print(new_files)
    for i in data_files:
        if i in new_files:
            print(i," is duplicate")
            flag=0
    
    if flag==1:
        print("No duplicate found")
        

     

    
```

```python
#create empty files
def create_files():
    dest_dir=input("Enter folder name Data or New: ")
    dest_folder=input("Enter folder name docs or pdf or jpg: ")
    num=int(input("Enter number of files you want to create: "))
    path="C:/Users/Administrator/Desktop"
    destination=os.path.join(path,dest_dir)
    destination1=os.path.join(destination,dest_folder)
    dir_list=os.listdir(destination1)
    print(dir_list)
    os.chdir(destination1)
    for i in range(num):
        open(f'test{i+1}.txt','w').close()

    print(os.listdir(destination1))
```

```python
flag=1
while(flag==1):
    ch=int(input("Enter 1 to compare number of files in Data and New\nEnter 2 to search a file\nEnter 3 to delete a file\nEnter 4 to copy a file\nEnter 5 to move a file\nEnter 6 to find duplicates\nEnter 7 to create files\n Enter 0 to exit"))
    if ch==1:
        compare_files()
    if ch==2:
        search_files()
    if ch==3:
        delete_files()
    if ch==4:
        copy_files()
    if ch==5:
        move_files()
    if ch==6:
        duplicate_files()
    if ch==7:
        create_files()
    if ch==0:
        flag=0
        print("EXIT")
        
```
