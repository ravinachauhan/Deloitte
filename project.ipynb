```python
'''you need to perform database operations using dictionary
where key is emp id and values are in a form of list having these details :-
Name, address, phone, gender, PAN card number, Department
create a menu driven program for adding new records, deleting records, filtering records based on some condition
ask user they want employees from which dept, phone numbers ending with 00, address containing any location like delhi
name having certain chars.
emp ids are unique
PAN card numbers are unique and of 10 chars
check for all entries so that no invalid data should be entered'''
```

```python
d={}
import sys
def user_input():
    
    try:
        print("Enter employee details below.")
        emp_id=int(input("Enter employee id: "))
        assert emp_id not in d.keys(),"already exist"
            
        name=input("Enter name: ")
        assert name.isalpha(),"name can have alphabets only"
        
        address=input("Enter address: ")
        
        phone=input("Enter phone number: ")
        assert len(phone)==10,'Phone number should have 10 digits'
        assert phone.isdigit(),'Phone number cannot have alphabets'
        
        gender=input("Enter gender M for male and F for Female: ")
        
        pan=input("Enter pan number: ") 
        assert len(pan)==10,'Pan number should have 10 digits.'
        
        department=input("Enter department: ")
        
    except AssertionError:
        print(sys.exc_info()[1])
    except ValueError:
        print("Employee id should be integer")
    else:
        d[emp_id]=[name,address,phone,gender,pan,department]


def add_emp():
    user_input()

def delete_emp():
    emp_id=int(input("Enter employee_id that you want to delete: "))
    if emp_id in d:
        d.pop(emp_id)
        print("Record deleted")
    else:
        print("Employee id not found")

def update_emp():
    try:
        emp_id=int(input("Enter employee id to update records: "))
        assert emp_id in d.keys(),"Employee id does not exist"
    except AssertionError:
        print(sys.exc_info()[1])
    except ValueError:
        print("Employee id should be integer")
    else:
        ch=int(input("Enter 1 to update name\nEnter 2 to update address\nEnter 3 to update phone number\nEnter 4 to update gender\nEnter 5 to update pan number\nEnter 6 to update department"))
        if ch==1:
               name=input("Enter new name: ")
               d[emp_id][0]=name
               print("Employee name updated")
        if ch==2:
               address=input("Enter new address: ")
               d[emp_id][1]=address
               print("Employee address updated")
        if ch==3:
               phone=input("Enter new phone number: ")
               d[emp_id][2]=phone
               print("Employee phone number updated")
        if ch==4:
               gender=input("Enter gender M for male and F for Female: ")
               d[emp_id][3]=gender
               print("Employee gender updated")
        if ch==5:
               pan=input("Enter pan number : ")
               d[emp_id][4]=pan
               print("Employee pan number updated")
        if ch==6:
               department=input("Enter new department")
               d[emp_id][5]=department
               print("Employee department updated")

def filter_emp():
    ch=int(input("Enter 1 to get records based on a specific name,\nEnter 2 to filter records based on address\nEnter 3 to filter records based on phone number\nEnter 4 to filter records based on gender\nEnter 5 to filter records based on pan number\nEnter 6 to filter records based on department\nEnter 7 to filter records with phone number ending with 00: "))
    if ch==1:
        name=input("Enter name to filter records: ")
        for i in d.values():
            if i[0]==name:
                print(i)
            else:
                print("No record found")
    if ch==2:
        address=input("Enter address to filter records: ")
        for i in d.values():
            if i[1]==address:
                print(i)
            else:
                print("No record found")
    if ch==3:
        phone=input("Enter phone number to filter records: ")
        for i in d.values():
            if i[2]==phone:
                print(i)
            else:
                print("No record found")
    if ch==4:
        gender=input("Enter gender to filter records: ")
        for i in d.values():
            if i[3]==gender:
                print(i)
            else:
                print("No record found")
    if ch==5:
        pan=input("Enter pan number to filter records: ")
        for i in d.values():
            if i[4]==pan:
                print(i)
            else:
                print("No record found")
    if ch==6:
        department=input("Enter department to filter records: ")
        for i in d.values():
            if i[5]==department:
                print(i)
            else:
                print("No record found")
    if ch==7:
        for i in d.values():
            if i[2][-2:]=='00':
                print(i)
            else:
                print("No record found")
                
flag=1
while(flag==1):
    ch=input("Enter 1 for adding employee\nEnter 2 for deleting employee\nEnter 3 for updating records.\nEnter 4 to filter records\nEnter 5 to view dictionary\nEnter 0 to exit: ")
    if int(ch)==1:
        add_emp()
    if int(ch)==2:
        delete_emp()
    if int(ch)==3:
        update_emp()
    if int(ch)==4:
        filter_emp()
    if int(ch)==5:
        if len(d)==0:
            print("Dictionary is empty")
        else:
            print(d)   
    if int(ch)==0:
        flag=0
        print("EXIT")
    
```
