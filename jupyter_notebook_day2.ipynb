```python
'''
take input from user first name and last name and concatenate to full name
 
in full name
1. count upper case letters
2.  count lower case letters
3.  count spaces
4. count special chars
5. count digits
6. find middle char of full name, if number of chars are even there will be 2 middle chars
   in case string length is odd, print only one middle char
7. find the length of whole full name
8. print reverse of a full name
9. print full name to lower case
 
'''
first_name=input("Enter first name")
last_name=input("Enter last name")
ucount=0
lcount=0
scount=0
dcount=0
spcount=0
full_name=first_name+" "+last_name
for ch in full_name:
    if ch.isupper():
        ucount=ucount+1
    if ch.islower():
        lcount=lcount+1
    if ch.isspace():
        scount=scount+1
    if ch.isdigit():
        dcount=dcount+1
    if ch=='@' or ch=='$':
        spcount=spcount+1

print("Upper case chracters: ",ucount)
print("lower case chracters: ",lcount)
print("digits : ",dcount)
print("spaces: ",scount)
print("special  chracters: ",spcount)
        
l=len(full_name)
m=l//2
if(m%2==0):
    print("Middle elements are ",full_name[m,m+2])
else:
    print("Middle element is ",full_name[m])

print("Reverse of full_name ",full_name[::-1])
print("lower case",full_name.lower())
```

```python
# take input from user any string and print smallest and largest ASCII value char in whole string
# ignore any number or space or special char from string and check only in upper and lower case letter

s=input("Enter the string : ")
max_val=ord(s[0])
min_val=ord(s[0])
for i in s[1:]:
    if i.isalpha():
        if ord(i)>max_val:
            max_val=ord(i)
            max_ch=i

        if ord(i)<min_val:
            min_val=ord(i)
            min_ch=i
        
print("smallest ASCII value char is ",min_ch)
print("largest ASCII value char is ",max_ch)
    
    
    
```

```python
'''
places = ['Delhi', 'Mumbai', 'Chennai', 'jaipur']
attractions =['red fort', 'gateway of india', 'marina beach', 'ameer fort']
 
1. print messages
Delhi has an attraction red fort
Mumbai has an attraction gateway of india
..
....
 
2. take input from user any place and print its attraction. search of place should be case insensitive.
if user enters CHENNAI he should get marina beach. if place is not found print not found
 

'''
flag=0
places = ['Delhi', 'Mumbai', 'Chennai', 'jaipur']
attractions =['red fort', 'gateway of india', 'marina beach', 'ameer fort']

for place, attraction in zip(places,attractions):
    print(place, " has an attraction ",attraction)
    
place=input("Enter any place: ")
for i in range(len(places)):
    if places[i].lower()==place.lower() :
        print(attractions[i])
        flag=1
if flag==0:
    print("Not found")
    
```

```python
import numpy
odd_list=[]
even_list=[]
marks=1
list1=[]
while(marks>0):
    marks=int(input("Enter Marks: "))
    if marks>0:
        list1.append(marks)

print(list1)    
print("average marks : ",numpy.mean(list1))
print("variance : ",numpy.var(list1))
print("stanadard deviation: ",numpy.std(list1))
print("corr : ",numpy.corrcoef(list1))

print("maximum marks : ",max(list1))
print("minimum marks : ",min(list1))

for mark in list1:
    if mark%2==0:
        even_list.append(mark)
    else:
        odd_list.append(mark)
        
print("even marks",even_list)
print("odd marks",odd_list)
```

```python
'''1. create a list having tuples having emp id and names of employees. keep on entereing data in list till user enters -999 for 
emp id. emp id cannot be negative. if user enters -ve emp id or -999 stop adding elements to a list.
your list should look like this
list1 = [(101, 'amit'), (102, 'sumit'), (103, 'AMIT')]
 
2. ask user any emp id and print name of that employee. if emp id not found print not found
3. ask user any name and create list of all employee ids having that name. eg if user enters amit, we should get
list as [101, 103]. search of name should be case insensitive.
 
4. find the name of employee having max number of characters.
5. find the name of employee whose name starts with A
6. find the name of employees whose name ends with n
 '''
emp=[]
id_list=[]
names_list=[]
flag=0
while( flag==0):
    emp_id=int(input("Enter id: "))
    if(emp_id<0 ):
        flag=1
    else:
        emp_name=input("Enter name: ")
        emp.append((emp_id,emp_name))
print(emp)       

emp_id=int(input("Enter emp_id: "))
for i in emp:
    if i[0]==emp_id:
        print(i[1])
        
    else:
        print("Not found")
        
name=input("Enter name : ")
for i in emp:
    if i[1]==name:
        id_list.append(i[0])
print("ids having same name ",id_list)




for i in emp:
    names_list.append(i[1])
print(names_list)
print("name with max length : ",max(names_list))
    
print("The names starting with A are : ")
for name in names_list:
    if name[0]=='A':
        print(name)
print("The names ending with n are")        
for name in names_list:
    if name[-1]=='n':
        print(name)
    else:
        print("Not found")
```
